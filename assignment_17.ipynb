```python
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import confusion_matrix
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sn
import pandas as pd
import numpy as np
%matplotlib inline
from sklearn.svm import SVC
```

```python
from sklearn.datasets import load_digits
```

```python
digit=load_digits()
x=digit.data
y=digit.target
```

```python
plt.imshow(x[4].reshape(8,8),cmap=plt.cm.gray)
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans()
kmeans
```

```python
kmeans = KMeans(n_clusters=10,max_iter=2000,random_state=1)
predict=kmeans.fit_predict(x)
```

```python
zero = np.zeros_like(predict)
```

```python
from scipy.stats import mode
for a in range(10):
  mask = (predict==a)
  zero[mask] = mode(y[mask])[0]
```

```python
from sklearn import metrics
metrics.accuracy_score(zero,y)
```

```python
confusion_matrix(zero,y)
```

```python
from sklearn.datasets import load_sample_image
china=load_sample_image('china.jpg')
plt.imshow(china)
```

```python
x,y,z=china.shape
x,y,z
```

```python
china1 = china.reshape(china.shape[0]*china.shape[1],china.shape[2])
china1.shape
```

```python
cluster=KMeans(n_clusters=7, random_state=42)
cluster.fit(china1)
cluster_centers = cluster.cluster_centers_
cluster_labels = cluster.labels_
```

```python
plt.figure(figsize = (15,8))
plt.imshow(cluster_centers[cluster_labels].reshape(x, y, z));
```
